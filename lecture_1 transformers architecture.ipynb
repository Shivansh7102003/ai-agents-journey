# Examples

## for T5 model

```python
from transformers import T5Tokenizer , T5ForConditionalGeneration

t5_model = T5ForConditionalGeneration.from_pretrained("t5-small")
t5_tokenizer = T5Tokenizer.from_pretrained("t5-small")


text = "translate english to german :the house is wonderful"
input_ids = t5_tokenizer.encode(text , return_tensors = "pt")

output = t5_model.generate(input_ids , max_length = 40)
translation = t5_tokenizer.decode(output[0] , skip_special_tokens = True)

print(f"input text: {text}")
print(f'translation: {translation}')
```

## for bert Models

```python
from transformers import pipeline
classifier = pipeline("sentiment-analysis" , model = "distilbert-base-uncased-finetuned-sst-2-english")

sentences = [
    "the movie i watched is amazing",
    "the product i bought is abolutely trash",
    "The capital of France is Paris."
]


print('bert sentiment analysis')

for sentence in sentences:
  result= classifier(sentence)[0]
  print(sentence)
  print(f'sentiment : {result['label']} , confidence: {result['score']:.4f}')
```

```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Load model and tokenizer
gpt_model = GPT2LMHeadModel.from_pretrained("gpt2")
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")

# Set padding token (GPT-2 doesn't have one by default)
gpt_model.config.pad_token_id = gpt_model.config.eos_token_id

# Prepare prompt
prompt = "Once upon a time in a distant land"
input_ids = tokenizer.encode(prompt, return_tensors="pt")

# Generate text
output = gpt_model.generate(
    input_ids,
    max_length=100,
    num_return_sequences=1,
    temperature=0.7,
    do_sample=True,
    top_p=0.9
)

# Decode output
generated_text = tokenizer.decode(output[0], skip_special_tokens=True)

print(prompt)
print(generated_text)
```
